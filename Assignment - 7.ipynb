<a id="Univariate"></a>
## <span style='color:Brown'> **4. Univariate Analysis** </span>

### <span style='color:Blue'> **Question - 3.Which gender count is more .**</span>
**Note :** Use bar graph

```python
#write observation and code here

gender_counts = df['sex'].value_counts()


gender_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)  
plt.show()

print(gender_counts)
```

### <span style='color:Blue'> **Question - 3.1.Which gender count is more display the percentage.**</span>
**Note :** Use pie chart

```python
#write observation and code here# 

gender_counts = df['sex'].value_counts()


plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.axis('equal') 

plt.show()


print(gender_counts)
```

### <span style='color:Blue'> **Question - 4.Calculate the counts of smoker columns & visualize using graph** </span>

```python
#write observation and code here

smoker_counts = df['smoker'].value_counts()


plt.figure(figsize=(6, 4))
smoker_counts.plot(kind='bar', color=['lightcoral', 'skyblue'], edgecolor='black')
plt.title('Smoker vs. Non-Smoker Count')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.show()

print(smoker_counts)
```

### <span style='color:Blue'> **Question - 4.1.Check the distribution of children column** </span>

```python
#write observation and code here

children_counts = df['children'].value_counts()


plt.figure(figsize=(8, 5))
children_counts.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Distribution of Children')
plt.xlabel('Number of Children')
plt.ylabel('Count')
plt.xticks(rotation=0)  
plt.show()

print(children_counts)
```

### <span style='color:Blue'> **Question - 4.2.Calculate the frequency count of Charges column** </span>

<a id="Bivariate"></a>
## <span style='color:Brown'> **5. Bivariate Analysis** </span>

### <span style='color:Blue'> **Question - 5.Which gender medical insurance charge is more** </span>
**Note :** Use any graph

```python
#write observation and code here

average_charges_by_gender = df.groupby('sex')['expenses'].mean()


plt.figure(figsize=(8, 6))
average_charges_by_gender.plot(kind='bar', color=['lightcoral', 'skyblue'], edgecolor='black')
plt.title('Average Medical Insurance Charges by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Charges')
plt.xticks(rotation=0)  
plt.show()


print(average_charges_by_gender)
```

### <span style='color:Blue'> **Question - 5.1.Which region has more insurance cost** </span>
**Note :** Use any graph

```python
#write observation and code here

average_cost_by_region = df.groupby('region')['expenses'].mean()

plt.figure(figsize=(10, 6))
average_cost_by_region.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Average Insurance Costs by Region')
plt.xlabel('Region')
plt.ylabel('Average Costs')
plt.xticks(rotation=0) 
plt.show()
```

### <span style='color:Blue'> **Question - 6.Visualize the medial insurance charge with respect to smoker columns** </span>
**Note :** Use any graph

```python
#write observation and code here

plt.figure(figsize=(8, 6))
plt.boxplot([df[df['smoker'] == 'yes']['expenses'], df[df['smoker'] == 'no']['expenses']], labels=['Smoker', 'Non-Smoker'])
plt.title('Medical Insurance Charges by Smoker Status')
plt.xlabel('Smoker Status')
plt.ylabel('Charges')
plt.show()
```

### <span style='color:Blue'> **Question - 6.1.Visualize insurance charge with respect to age group** </span>
**Note :** Use any graph

```python
#write observation and code here

plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['expenses'], c='blue', alpha=0.6)
plt.title('Insurance Charges vs. Age')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.grid(True)
plt.show()
```

<a id="Multivariate"></a>
## <span style='color:Brown'> **6. Multivariate Analysis** </span>

### <span style='color:Blue'> **Question - 7.Which age group female has more insurance cost** </span>


```python
#write observation and code here

plt.figure(figsize=(10, 6))
plt.scatter(df[df['sex'] == 'female']['age'], df[df['sex'] == 'female']['expenses'], c='blue', label='Female', alpha=0.6)
plt.scatter(df[df['sex'] == 'male']['age'], df[df['sex'] == 'male']['expenses'], c='red', label='Male', alpha=0.6)
plt.title('Insurance Charges vs. Age (Females vs. Males)')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.legend()
plt.grid(True)
plt.show()
```

### <span style='color:Blue'> **Question - 8.Which region males has more insurance cost** </span>


```python
#write observation and code here

plt.figure(figsize=(10, 6))
plt.scatter(df[df['sex'] == 'female']['region'], df[df['sex'] == 'female']['expenses'], c='blue', label='Female', alpha=0.6)
plt.scatter(df[df['sex'] == 'male']['region'], df[df['sex'] == 'male']['expenses'], c='red', label='Male', alpha=0.6)
plt.title('Insurance Charges vs. Region (Females vs. Males)')
plt.xlabel('Region')
plt.ylabel('Insurance Charges')
plt.legend()
plt.xticks(rotation=45)  
plt.grid(True)
plt.show()
```

### <span style='color:Blue'> **Question - 9.Visualize the columns smoker & insurange charges with respect to gender** </span>


```python
#write observation and code here

plt.figure(figsize=(10, 6))
colors = {'female': 'blue', 'male': 'red'}
plt.scatter(df['smoker'], df['expenses'], c=df['sex'].map(colors), alpha=0.6)
plt.title('Smoker vs. Insurance Charges (colored by Gender)')
plt.xlabel('Smoker Status')
plt.ylabel('Insurance Charges')
plt.legend(df['sex'].unique()) 
plt.grid(True)
plt.show()
```

### <span style='color:Blue'> **Question - 10.Check the relationship between age & insurance charge with respect to gender** </span>


```python
#write observation and code here

plt.figure(figsize=(10, 6))
colors = {'female': 'blue', 'male': 'red'}
plt.scatter(df[df['sex'] == 'female']['age'], df[df['sex'] == 'female']['expenses'], c='blue', label='Female', alpha=0.6)
plt.scatter(df[df['sex'] == 'male']['age'], df[df['sex'] == 'male']['expenses'], c='red', label='Male', alpha=0.6)
plt.title('Age vs. Insurance Charges (colored by Gender)')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.legend()
plt.grid(True)
plt.show()
```
