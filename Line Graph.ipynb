```python
import matplotlib.pyplot as plt
```

```python
s1 = [1,2,3,4,5]
s2 = [10,6,15,9,7]
```

```python
plt.plot(s1,s2)
plt.show()
```

```python
s3= [10,5,16,3,21]
```

```python
plt.plot(s1,s2,label ='line A' ,color = 'red',marker ='o',markersize =5,linestyle ='dashed',linewidth =3)
plt.plot(s1,s3,label ='line B',color ='Green',marker ='*',markersize =8,linewidth =3)
plt.legend()
plt.grid()
plt.title('Line Graph')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')

plt.show()
```
