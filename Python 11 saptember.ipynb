#### pie

- A pie chart in Python is a circular graph that divides a categorical dataset into slices, where each slice's size corresponds to the proportion or percentage of a specific category within the whole dataset.

```python
plt.figure(figsize-(6,4))
df['smoker'].value_counts().plot(kind='pie',autopct='%0.3f%%')
plt.show()
```

### KDE(Kernel Density Estimation)

KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
plt.figure(figsize=(5,3))
df['total_bill'].plot(kind='kde',color='g')
plt.grid()
plt.xlabel('Total bill ($)')
plt.show()
```

### kdeplot

```python
#kdeplot
plt.figure(figsize=(5,3))
sns.kdeplot(data= df,x = 'total_bill',fill= True)
plt.show()
```

```python
#kdeplot
plt.figure(figsize=(5,3))
sns.kdeplot(data= df,y = 'total_bill',fill= True)
plt.show()
```

### violin plot 
- is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. 

- violinplot=kde+boxplot

```python
sns.violinplot(data =df, y ='size')
plt.grid()
plt .show()
```

histplot: The histplot function is primarily focused on creating histograms. While it can include a KDE overlay, its primary purpose is to display the frequency distribution of data in histogram form

```python
plt.hist(df['tip'],edgecolor='k')
plt.xlabel('Tip ($)')
plt.ylabel('value counts of tips')
plt.show()
    
```

```python
plt.hist(df['tip'],edgecolor='k',color='g',bins=50)
plt.xlabel('Tip ($)')
plt.ylabel('value counts of tips')
plt.show()
    
```

```python
plt.hist(df['tip'],edgecolor='r',color='c',bins=50)
plt.xlabel('Tip ($)')
plt.ylabel('value counts of tips')
plt.show()
```

distplot: The distplot function was designed to create combined histograms and kernel density plots. It allowed you to visualize the distribution of a dataset using both histogram bars and a smoothed KDE line.
