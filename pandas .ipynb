```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df1  = DataFrame({'Key': list('XYXYYZXY'),'Dataset1':[1,2,3,4,5,6,7,8]})
```

```python
df1
```

```python
df2 = DataFrame({'Key':list('QYZ'),'Dataset2':[1,2,3]})
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, on ='Key')
```

```python
pd.merge(df1,df2,on ='Key',how ='inner')
```

```python
pd.merge(df1,df2,on ='Key',how ='outer')
```

```python
pd.merge(df1,df2,on ='Key',how ='left')
```

```python
pd.merge(df1,df2,on ='Key',how ='right')
```
