### Advance Scatterplot in python33

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import pandas as pd
```

```python
df1 = sns.load_dataset('tips')
```

```python
df1
```

```python
df1.head()
```

```python
plt.scatter(df1['total_bill'],df1['tip'])
plt.title('Total Bill Vs Tip given')
plt.xlabel('Total Bill')
plt.ylabel('Tip Given')
plt.show()
```

```python
sns.scatterplot(x='total_bill',y ='tip',data =df1)
plt.show()
```

```python
sns.scatterplot
```

```python
sns.scatterplot(x ='total_bill',y='tip',hue ='sex',data =df1)#hue is uesd for sub segment of data
plt.show()
```

```python
sns.scatterplot(x ='total_bill',y='tip',hue ='day',data =df1)
plt.title('Total Bill Vs Tip given')
plt.xlabel('Total Bill')
plt.ylabel('Tip Given')
plt.show()
```

```python
df1.plot(kind= 'scatter',x='total_bill',y ='tip')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data =df1)
plt.show()
```
