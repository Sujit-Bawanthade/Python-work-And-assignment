```python
sns.histplot(data= df1, bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Data Visualization: Gender Distribution
# Gender Distribution
plt.figure(figsize=(6, 6))
sns.countplot(data=df1, x='Gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Data Visualization: Marital Status Distribution
marital_status_counts = df1['Marital_Status'].value_counts()
plt.figure(figsize=(6, 6))
plt.bar(marital_status_counts.index, marital_status_counts.values)
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status (0 - Unmarried, 1 - Married)')
plt.ylabel('Count')
plt.show()
```

```python
# Calculate Minimum and Maximum Age
min_age = df1['Age'].min()
max_age = df1['Age'].max()

print(f"Minimum Age: {min_age}")
print(f"Maximum Age: {max_age}")
```

```python
# Total Amount Spent by Gender
plt.figure(figsize=(8, 6))
sns.barplot(data=df1, x='Gender', y='Amount', estimator=sum)
plt.title('Total Amount Spent by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Marital Status Distribution
plt.figure(figsize=(6, 6))
sns.countplot(data=df1, x='Marital_Status')
plt.title('Marital Status Distribution')
plt.xlabel('Marital Status (0 - Unmarried, 1 - Married)')
plt.ylabel('Count')
plt.show()
```

```python
# Age Group Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df1, x='Age Group', order=['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+'])
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram for Age
plt.figure(figsize=(8, 6))
plt.hist(df1['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Histogram for Amount
plt.figure(figsize=(8, 6))
plt.hist(df1['Amount'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Bar Chart for Product Category Counts
product_category_counts = df1['Product_Category'].value_counts()
plt.bar(product_category_counts.index, product_category_counts.values, color='skyblue')
plt.title('Product Category Counts')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Pie Chart for Product Category Distribution
plt.figure(figsize=(8, 8))
plt.pie(product_category_counts, labels=product_category_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(product_category_counts))))
plt.title('Product Category Distribution')
plt.axis('equal')
plt.show()
```

```python
# Sort the data by 'Age' for a sequential representation
df_sorted = df1.sort_values(by='Age')

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(df_sorted['Age'], marker='o', linestyle='-')
plt.title('Age Distribution (Line Chart)')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
